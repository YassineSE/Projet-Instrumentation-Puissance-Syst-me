```python
import matplotlib.pyplot as plt
import csv
```

```python

# Define the file path
csv_file_path = '4003_resistance_vs_temperature.csv'

# Initialize an empty dictionary with keys 'Temperature' and 'Resistance'
data_dict = {
    "Temperature": [],
    "Resistance": []
}

# Open the CSV file and load data into the dictionary
with open(csv_file_path, mode='r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        # Append temperature and resistance values to their respective lists
        data_dict["Temperature"].append(float(row['Temperature (°C)']))
        data_dict["Resistance"].append(float(row['Resistance (RT/R25) 4003']))

# Print the dictionary
print(data_dict)
```

```python
R_25 = 47
data_dict["Resistance"] = [R_25 * i for i in data_dict["Resistance"]]
```

```python
plt.plot(data_dict["Temperature"], data_dict["Resistance"])
plt.title("Résistance en Fonction de la température")
```

```python

plt.scatter(data_dict["Temperature"], data_dict["Resistance"])
plt.title("Résistance en Fonction de la température (scatter)")
```
